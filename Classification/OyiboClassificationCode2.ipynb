```python
import numpy as np
import pandas as pd
```

<h1>Training Dataset 2</h1>

```python
#loading TrainData1.txt
TD2 = pd.read_csv(r"/Users/bada/Downloads/48506850Project/TrainData2.txt", delimiter='\s+', header = None)
TD2_features = TD2
TD2_labels = pd.read_csv(r"/Users/bada/Downloads/48506850Project/TrainLabel2.txt", delimiter='\t', header = None)
```

```python
TD2_dataset = pd.DataFrame(TD2)
TD2_labels = pd.DataFrame(TD2_labels)
```

<h2>Handling Missing Values</h2>
A missing entry is filled by 1.00000000000000e+99

```python
#How many missing values are there?
TD2_missing_entry_val= [1.00000000000000e+99]
TD2_missingvals = TD2_dataset.isin(TD2_missing_entry_val).sum().sum()
TD2_missingvals
```

```python
#How many values are missing in each column
TD2_dataset.isin(TD2_missing_entry_val).sum()
```

```python
from sklearn.model_selection import train_test_split
#Splitting the data. The Training data will be X and our Training label will be y
X = TD2_features
y = TD2_labels
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8 ,test_size=0.2, random_state=42)
```

```python
unique_classes, counts = np.unique(y_train, return_counts=True)
print("Class distribution:")
for cls, count in zip(unique_classes, counts):
    print(f"Class {cls}: {count} samples")
```

<h3>Scaling</h3>

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
#fit = learns the parameters ( mean and standard deviation)

scaled = scaler.fit_transform(X)
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm


rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_y_pred = rf.predict(X_test)
print('\nAccuracy Random Forest:', metrics.accuracy_score(y_test, rf_y_pred))

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_y_pred = knn.predict(X_test)
print('\nAccuracy KNN:', metrics.accuracy_score(y_test, knn_y_pred))

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
dtc_y_pred = dtc.predict(X_test)
print('\nAccuracy DTC:', metrics.accuracy_score(y_test, dtc_y_pred))

gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb_y_pred = gnb.predict(X_test)
print('\nAccuracy GaussianNB:', metrics.accuracy_score(y_test, gnb_y_pred))

clf = svm.SVC(kernel='linear') 
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('\nAccuracy SVM:',metrics.accuracy_score(y_test, y_pred))
```

<h3>Which Model had the highest accuracy?</h3>
<b>Support Vector Machine</b> and <b>Random Forest</b> have the highest accuracy scores of <b>95%</b>

<h2> Hypertuning: What are the Best Paramters for Our Chosen Model(s)?</h2>
Since both <b>Support Vector Machine</b> and <b>Random Forest</b> have an accuracy score of <b>95%</b>, We will hypertune to see which paramaters work best for SVM & RF

```python
from sklearn.model_selection import GridSearchCV

param_grid_svm = {
            'C': [0.1, 1, 10, 100, 1000],  
            'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
            'kernel': ['linear']}

grid_search_svm = GridSearchCV(clf, param_grid_svm, scoring='accuracy', refit = True, verbose = 3, n_jobs=-1) 
  
# fitting the model for grid search 
grid_search_svm.fit(X_train, y_train)

# print best parameter after tuning 
grid_search_svm.best_params_
```

```python
param_grid_rf = {
            'n_estimators': [500, 700, 1000],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 10],
            'min_samples_leaf': [1, 5]}

grid_search_rf = GridSearchCV(rf, param_grid_rf, scoring='accuracy', refit = True, verbose = 3, n_jobs=-1) 
  
# fitting the model for grid search 
grid_search_rf.fit(X_train, y_train)

grid_search_rf.best_params_
```

<h2>The Best Parameters for SVM</h2>
C': 0.1, 'gamma': 1, 'kernel': 'linear'

<h2>The Best Parameters for RF</h2>
'max_depth': 20,
<br>'min_samples_leaf': 1,

```python
svm_updated = svm.SVC(C= 0.1, gamma = 1, kernel='linear')
svm_updated.fit(X_train, y_train)
y_pred_USVM = svm_updated.predict(X_test)
print("Test Accuracy:", metrics.accuracy_score(y_test, y_pred_USVM))
```

```python
rf_updated = RandomForestClassifier(max_depth = 20,
                 min_samples_leaf= 1,
                 min_samples_split = 2,
                 n_estimators = 1000)
rf_updated.fit(X_train, y_train)
y_pred_URF = rf_updated.predict(X_test)
print("Test Accuracy:", metrics.accuracy_score(y_test, y_pred_URF))
```
